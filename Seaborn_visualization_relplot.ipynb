```python
import seaborn as sns
sns.__version__
sns.set_style(style="darkgrid")
```

```python
Tips = sns.load_dataset("tips")
Tips.tail()
```

```python
Tips.info()
```

```python
Tips["sex"].value_counts()
```

```python
Tips['smoker'].value_counts()
```

```python
Tips["day"].value_counts()
```

```python
Tips["time"].value_counts()
```

```python
sns.regplot(
    data= Tips,
    x = "total_bill",
    y = "tip"
)
```

```python
sns.relplot(
    data= Tips,
    x = "total_bill",
    y = "tip",
    hue = "day"
)
```

```python
sns.relplot(
    data = Tips,
    x = "total_bill",
    y = "tip",
    hue = "day",
    col = "day"
)
```

```python
sns.relplot(
    data = Tips,
    x= "total_bill",
    y = "tip",
    col = "time",
    hue = "sex"
)
```

```python
sns.relplot(
    data = Tips,
    x = "total_bill",
    y = "tip",
    col = "time",
    hue = "sex",
    style= "smoker"
)
```
