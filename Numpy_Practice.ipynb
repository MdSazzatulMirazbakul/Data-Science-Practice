Numpy Array Iteration

```python
print(np_array)
```

```python
"""Iterate by value"""
for val in np_array:
    print(val)
```

```python
"""Iterate by index"""
for i in range(len(np_array)):
    print(np_array[i])
```

```python
"""Iterate by value and Index"""
for i, val in enumerate(np_array):
    print(f"Index: {i} Value: {val}")
```

```python
New_array = np.arange(
    start= 40, stop= 120, step= 7,
)
print(New_array)
for index, value in enumerate(New_array):
    print(f"Index: {index}, Value: {value}")
```

```python
random_array = np.array([8, 3, 15, 59, 38, 5, 23, 11, 30])
sorted_random_array = np.sort(random_array)
reversed_sorted_random_array = np.sort(random_array)[::-1]
print(sorted_random_array)
print(reversed_sorted_random_array)
```

```python
print(np.argmax(random_array))
print(np.argmin(random_array))
```
